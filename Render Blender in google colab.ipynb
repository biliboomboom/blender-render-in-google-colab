```python
#8
#渲染启动行！
#渲染单帧图片模式

blenderFolderLocation = '/content/drive/MyDrive/blendin/'#确认这是你blender应用程序所在路径
outputFolderLocation = '/content/drive/MyDrive/blenderout/'#确认这是你需要保存渲染结果的路径

blenderFileName = input("输入你的blender工程文件名 (需要加上.blend的后缀)，例如：甜甜圈.blender : ")
form = input("需要渲染的帧位：")

tempFile = blenderFileName.split('.')
outputFileName = tempFile[0]

outputFolderName = outputFolderLocation + outputFileName

blenderFile = blenderFolderLocation + blenderFileName
outputFile = outputFolderName + '/' + outputFileName + "_"

!mkdir -p $outputFolderName

#Run Blender
#（可选）如果获得支持 OPTIX 的 GPU，则可以更改为行尾。Tesla K80 GPU 不支持 OPTIX。     copy: CUDA OPTIX
!sudo /content/drive/MyDrive/blender3linuxbate/blender -b $blenderFile -noaudio -o $outputFile  -E 'CYCLES' -f $form -F 'PNG' -- --cycles-device OPTIX
#!sudo /content/drive/MyDrive/blender3linuxbate/blender -b $filename -noaudio -- --cycles-device CUDA  -E 'CYCLES' -f 1 -F 'PNG'
```
