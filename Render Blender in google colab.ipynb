```python
# 4
#每次渲染前需执行

#赋予当前用户执行权限
!chmod +x /content/drive/MyDrive/blenderFile_Linux/blender
```

```python
# 5
# 每渲染一个新文件需执行一次

#输入你的blender工程文件全名（带后缀）

filenamere = input("输入你的blender工程文件全名(.blend“带后缀”):")
filenameroad = '/content/drive/MyDrive/in_blender/'#确认这是你的blender工程文件路径
filename = filenameroad + filenamere
```

```python
# 6
#渲染启动行！
#渲染单帧模式!

blenderFolderLocation = '/content/drive/MyDrive/in_blender/'#确认这是你blender应用程序所在路径
outputFolderLocation = '/content/drive/MyDrive/out_blender/'#确认这是你需要保存渲染结果的路径

#blenderFileName = input("输入你的blender工程文件名 (需要加上.blend的后缀)，例如：甜甜圈.blender : ")
form = input("需要渲染的帧位：")

tempFile = filenamere.split('.')
outputFileName = tempFile[0]

outputFolderName = outputFolderLocation + outputFileName

blenderFile = blenderFolderLocation + filenamere
outputFile = outputFolderName + '/' + outputFileName + "_"

!mkdir -p $outputFolderName

#Run Blender
#（可选）如果获得支持 OPTIX 的 GPU，则可以更改为行尾。Tesla K80 GPU 不支持 OPTIX。     copy: CUDA OPTIX
!sudo /content/drive/MyDrive/blenderFile_Linux/blender -b $blenderFile -noaudio -o $outputFile  -E 'CYCLES' -f $form -F 'PNG' -- --cycles-device OPTIX #（t4）
#!sudo /content/drive/MyDrive/blenderFile_Linux/blender -b $blenderFile -o $outputFile -E CYCLES  -x 1 -f  $form -F 'PNG' #（k80）
```

```python
# 7
#渲染多帧动画模式

blenderFolderLocation = '/content/drive/MyDrive/in_blender/'
outputFolderLocation = '/content/drive/MyDrive/out_blender/'

#blenderFileName = input("Enter the name of the Blender File (with .blend extension) : ")

tempFile = filenamere.split('.')
outputFileName = tempFile[0]

outputFolderName = outputFolderLocation + outputFileName

blenderFile = blenderFolderLocation + filenamere
outputFile = outputFolderName + '/' + outputFileName + "_"

!mkdir -p $outputFolderName

#设置开始和结束帧
startFrame = int(input("\nEnter the Starting Frame : "))
endFrame = int(input("\nEnter the Ending Frame : "))

for frame in range(startFrame, endFrame+1):
  #Run Blender
  !sudo /content/drive/MyDrive/blenderFile_Linux/blender -b $blenderFile -noaudio -o $outputFile  -E 'CYCLES' -f $frame -F 'PNG' -- --cycles-device OPTIX #（t4）
  #!sudo /content/drive/MyDrive/blenderFile_Linux/blender -b $blenderFile -o $outputFile -E CYCLES  -x 1 -f  $form -F 'PNG' #（k80）
```
